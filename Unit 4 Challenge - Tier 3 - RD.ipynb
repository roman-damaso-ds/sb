**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Creating function that returns a ratio of average price in 2020 to average price in 2000
def create_price_ratio(d):
    y2000 = float(d['Average_price'][d['Year']==2000])
    y2020 = float(d['Average_price'][d['Year']==2020])
    ratio = [y2020/y2000]
    return ratio
```

```python
#Testing out function on a borough
create_price_ratio(dfg[dfg['London_Boroughs']=='City of London'])
```

```python
#Will scale the out by creating a dictionary and then a for loop to get that ratio from all the other boroughs
#First the dictionary
final = {}
```

```python
#Now the for loop
for b in dfg['London_Boroughs'].unique():
    #Replicating the test function's format
    borough = dfg[dfg['London_Boroughs']==b]
    #Filling it in the dictionary
    final[b] = create_price_ratio(borough)
print(final)
    
```

```python
#Need to now convert dictionary into a dataframe by assigning it as argument to dataframe method
df_ratios = pd.DataFrame(final)
```

```python
#checking out df_ratios
df_ratios.head()
```

```python
#Transpose data and reset index, check it out via head
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
#Fix the headers from index and 0 to London_Boroughs and 2020, respectively; then check the head
df_ratios.rename(columns = {'index':'Borough',0:'2020'}, inplace=True)
df_ratios.head()
```

```python
#Sort them descending and select top 10
top10 = df_ratios.sort_values(by='2020',ascending=False).head(10)
print(top10)
```

```python
#Sort them ascending and select bottom 10
bottom10 = df_ratios.sort_values(by='2020',ascending=True).head(10)
print(bottom10)
```

```python
#Plot boroughs that have seen greatest change in price
#Making variable ax. Assigning it result of filtered top10 on borough and 2020, calling plot method, with kind bar parameter
ax = top10[['Borough','2020']].plot(kind='bar')
ax.set_xticklabels(top10.Borough)
```
